## HOUR DAY TRANSFORMATION

```python
import datetime
import calendar
```

```python
uber_data['START_DATE*'] = pd.to_datetime(uber_data['START_DATE*'], format="%m/%d/%Y %H:%M") 
uber_data['END_DATE*'] = pd.to_datetime(uber_data['END_DATE*'], format="%m/%d/%Y %H:%M")
```

```python
hour=[]  #empty list
day=[]
dayofweek=[]
month=[]
weekday=[]
for x in uber_data['START_DATE*']:
    hour.append(x.hour) #adding/appending the values to above empty list
    day.append(x.day)
    dayofweek.append(x.dayofweek)
    month.append(x.month)
    weekday.append(calendar.day_name[dayofweek[-1]])
uber_data['HOUR']=hour #creatig columns
uber_data['DAY']=day
uber_data['DAY_OF_WEEK']=dayofweek
uber_data['MONTH']=month
uber_data['WEEKDAY']=weekday
```

```python
uber_data.head() #updated dataset
```

```python
uber_data['HOUR'].value_counts().plot(kind='bar',figsize=(10,5),color='red') #trips per hour
```

```python
uber_data['WEEKDAY'].value_counts().plot(kind='bar',figsize=(10,5),color='black') #trips per day of week
```

```python
uber_data['DAY'].value_counts().plot(kind='bar',figsize=(10,5),color='blue') #trips per day of the month
```

```python
uber_data['MONTH'].value_counts().plot(kind='bar',figsize=(10,5),color='green') #trips in month, month 12 has highest number of trips as depicted in the figure
```

```python
uber_data.groupby('PURPOSE*').mean().plot(kind='bar',figsize=(10,5)) #overall analysis
```
